```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/VictimasVS_202509.xlsx'
data = pd.read_excel(file_path)
```

```python
columnas = ["Etnia", "Tipo de Población Vulnerable"]
data[columnas]=data[columnas]. fillna("NO APLICA")
columna=["Edad"]
data[columna]=data[columna].fillna ("NO REPORTADA")
```

```python
data_filtrada_años= data[(data['Año'] >= 2015) & (data['Año'] <= 2025)]
```

```python
remove_rows= ['ID Caso','Código DANE de Municipio','Mes', 'Día', 'ID Persona','Calidad de la Víctima o la Baja', 'Militante Político','Fuerza o Grupo Armado Organizado al que Pertenece el Combatiente',	'Descripción Fuerza o Grupo Armado Organizado al que Pertenece el Combatiente','Latitud',	'Longitud']
data = data.drop (remove_rows, axis =1 )
```

```python
Media = data_filtrada_años["Año"].mean()
Moda = data_filtrada_años["Año"].mode()[0]
Mediana = data_filtrada_años["Año"].median()
print(f"Media: {Media}")
print(f"Moda: {Moda}")
print(f"Mediana: {Mediana}")

# Plotting the histogram with mean, median, and mode
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
data_filtrada_años['Año'].hist(bins=range(int(data_filtrada_años['Año'].min()), int(data_filtrada_años['Año'].max()) + 2), edgecolor='black')
plt.axvline(Media, color='red', linestyle='dashed', linewidth=2, label=f'Media: {Media:.2f}')
plt.axvline(Moda, color='green', linestyle='dashed', linewidth=2, label=f'Moda: {Moda}')
plt.axvline(Mediana, color='purple', linestyle='dashed', linewidth=2, label=f'Mediana: {Mediana:.2f}')
plt.title('Distribución del Año con Media, Moda y Mediana')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Mean and Median are not applicable for categorical data.
Moda = data_filtrada_años["Municipio"].mode()[0]
print(f"Moda: {Moda}")

# Plotting the distribution of 'Etnia' using a bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
data_filtrada_años['Municipio'].value_counts().plot(kind='line')
plt.title('Distribución de Municipio')
plt.xlabel('Municipio')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
